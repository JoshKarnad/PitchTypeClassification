```python
df = pd.read_csv('modelingv2.csv')
```

```python
# Creating histograms to see if I can get a more normalized distribution for these variables

def make_plots(feature):
    fig = plt.figure(figsize=(8,8))

    fig.add_subplot(221)
    plt.hist(df[feature])
    plt.title(feature + '- Raw')

    fig.add_subplot(222)
    plt.hist(np.sqrt(df[feature]))
    plt.title(feature + '- Square root')

    fig.add_subplot(223)
    plt.hist((df[feature]*df[feature]))
    plt.title(feature + '- Squared')
    
    fig.add_subplot(224)
    plt.hist(np.log(df[feature].dropna()))
    plt.title(feature + '- Log')
    
    plt.show()
```
